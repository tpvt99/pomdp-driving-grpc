Here I plot the pid plot with respect to time.
x-axis is time and y-axis is the response

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
# First method: Scaling PID gains: new_gains = old_gains * (new_freq/old_freq). Using old
input_file = '/home/phong/driving_data/del/cv0_1/result/joint_pomdp_drive_mode/beijing/pomdp_search_log-0_0_pid-57_r-3945420.txt'
```

```python
# We read files
def return_command_and_current_vel(file_path):
    command_vel = []
    current_vel = []
    with open(file_path, 'r') as f:
        lines = f.readlines()
        start_reading = False
        for line in lines:
            if "steer/acc/speed" in line:
                speed = float(line.split('/')[-1])
                command_vel.append(speed)
                start_reading = True
            if start_reading and "car pos / heading" in line:
                speed = float(line.split('/')[4].split()[0])
                current_vel.append(speed)
    return command_vel, current_vel
```

```python
# The plot above shows overshooting behavior. which is not really good.

command_vel, current_vel = return_command_and_current_vel(input_file)
x = np.arange(len(command_vel))
x_current = np.arange(len(current_vel)-1)

figure(figsize=(30, 10), dpi=80)
plt.plot(x, command_vel, label='command_vel', color='red')
plt.plot(x_current, current_vel[1:], label='current_vel', color='blue')
plt.legend()
plt.show()
```
